**Bias and variance**


**Bias** = error from wrong assumptions

Bias measures how far your model‚Äôs average prediction is from the true value.

1. High bias ‚Üí oversimplified model

2. The model can‚Äôt capture patterns in data

3. It underfits

**Examples**:

1. Using a straight line to fit highly nonlinear data

2. Linear regression on complex fraud patterns


**Symptoms**:

1. Low training accuracy

2. Low test accuracy

3. Model predictions look ‚Äúsimple‚Äù.






 **Variance** = error from sensitivity to training data

Variance measures how much your model‚Äôs predictions change when the training data changes.


1. High variance ‚Üí overly complex model


2. The model memorizes noise


3. It overfits


**Examples**


1. Very deep decision trees


2. Neural networks without regularization


**Symptoms**


1. Very high training accuracy


2. Poor test accuracy


3. Model reacts too much to small fluctuations
